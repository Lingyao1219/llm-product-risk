#### Sample Processing

```python
sample_df = pd.read_csv('data_collection/samples.csv')
sample_df['post_id'] = range(len(sample_df))
sample_df = sample_df.sample(n=10)
```

```python
def analyze_dataframe(df, text_column="text", sample_size=None):
    """
    Analyze texts in a DataFrame using the Analyzer.
    """
    analyzer = Analyzer()
    if sample_size is not None and sample_size < len(df):
        sample_df = df.sample(sample_size).copy()
    else:
        sample_df = df.copy()
    
    system_prompt = get_system_prompt()
    task_prompt = get_task_prompt()
    
    processed_df = analyzer.annotate_dataset(
        dataset=sample_df,
        text_column=text_column,
        system_prompt=system_prompt,
        task_prompt=task_prompt
    )

    return processed_df

processed_df = analyze_dataframe(sample_df)
processed_df
```

### Files Compiling

```python
batch_files = glob.glob('annotation/batch_*.csv')

dataframes = []
for file in batch_files:
    df = pd.read_csv(file)
    dataframes.append(df)

compiled_df = pd.concat(dataframes, ignore_index=True, sort=False)
compiled_df.to_csv('samples_output.csv', index=False)
```
